```python
#EDA

import pandas as pd 
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("./datasets/dataset.csv")
columns_n = df.columns.tolist()

print(df.describe())
print(df.info())
```

```python
#nan ?

print(df.isna().sum())

sns.heatmap(df.isna(), cbar=False)
df_clean = df.dropna().reset_index(drop=True)

```

```python
#distributions of game per year
df_clean.year.hist(bins=20)
plt.xlabel('Year')
plt.ylabel('Number of Games Released')
plt.show()

#rating distribution

df_clean.metacritic_rating.hist(bins=20)
plt.ylabel("Metacritic rating")
plt.show("N of games")

```

```python
#Relationship of ratings and reviews

plt.scatter(df_clean.reviewer_rating, df_clean.metacritic_rating)
plt.title("Relationship of M.ratings and reviewers Rating")
plt.ylabel("Metacritic")
plt.xlabel("Reviewer")
plt.show()
```

```python
#Relationship of Reviewers critic vs main completion time 

plt.scatter(df_clean.reviewer_rating, df_clean.to_beat_main)
plt.xlabel("Review")
plt.ylabel("time to complete (Days)")
```
